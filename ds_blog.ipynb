The **Country** column is straightfoward, but the **HaveWorkedLanguage** column has mutiple entries. We need to find a way to process this information and relate it to each country. Below is the list of **HaveWorkedLanguage** possible values retrieved form the survey documentaion. I will write a function to process the data by a single country at a time, and calculate the number of respondents that indicated they have worked with each language.

```python
def get_all_values(df, column, separator):
    working_df = pd.DataFrame(df, columns=[column]).dropna()
    results = []
    for index, row in working_df.iterrows():
        entry = row[column] 
        values = entry.split("; ")
        for val in values:
            if val not in results:
                results.append(val)
            
    return results
```

```python
possible_languages = get_all_values(df, "HaveWorkedLanguage", "; ")
```

```python
def get_lang_count_for_country(df, country):
    #df2 = df[df["Country"]==country]
    df2 = pd.DataFrame(df[df["Country"]==country], columns=["HaveWorkedLanguage"]).dropna()
    #results = {i: 0 for i in possible_languages}
    results = dict.fromkeys(possible_languages, 0)
    for index, row in df2.iterrows():
        entry = row["HaveWorkedLanguage"] 
        #print(entry)
        #if not np.isnan(entry):
        values = entry.split("; ")
        for val in values:
            results[val] += 1
            
    results =  pd.DataFrame(pd.Series(results)).reset_index()
    results.columns = ["Language", "Count"]
    results.sort_values("Count", ascending=False, inplace=True)
    results.reset_index()
    return results
```

# Question 1 Discussion

This provides a way of determinig the popular language for each country. Above we took a look at the top 5 languages in three countries, Germany, The US and Nigeria. JavaScript is the most popular languange in all three countries, Java, C# and SQL feature in all three. It is worth noting that these ranking are relative to each country, and is not representative of the total number of users, for example, although C# only has 953 users in Germany, it is still higher in position than C# in the US with 3100 users amonsgt the respondents. A line of thought to explore popularity by country is to ask which is the language that is most popular in largest number of countries, and which is second most popuplar etc. 

To achieve this, we can create a sort of popularity index by language depending on it's raking. Let's explore this in the cells below. Let's begin by finding the most popular language only.

```python
#Let's let's modify the get_lang_count_for_country function to return only the top x languages in a set

from operator import itemgetter
def get_top_k_langs_for_country(df, country, k=1, kth_item_only = False):
    df2 = pd.DataFrame(df[df["Country"]==country], columns=["HaveWorkedLanguage"]).dropna()
    results = dict.fromkeys(possible_languages, 0)
    for index, row in df2.iterrows():
        entry = row["HaveWorkedLanguage"] 
        values = entry.split("; ")
        for val in values:
            results[val] += 1
            
    if kth_item_only:
        sorted_results = dict(sorted(results.items(), key = itemgetter(1), reverse = True)[k-1:k])
    else:
        sorted_results = dict(sorted(results.items(), key = itemgetter(1), reverse = True)[:k])
    return sorted_results 

        
```

```python
print(get_top_k_langs_for_country(df, "Nigeria", 3))
print(get_top_k_langs_for_country(df, "Nigeria", 3, True))
```

```python
# Let's create a function to tally most popular language for all countries
#let's define a function for that

def get_popularity_index_for_languages(df, langs_of_interest_list, search_countries_list, position = 1, as_dataframe=False):
    results = dict.fromkeys(langs_of_interest_list, 0)
    for c in search_countries_list:
        v = list(get_top_k_langs_for_country(df, c, position, False).keys())[position-1]
        #print(v)
        if v in langs_of_interest_list:
            results[v] += 1
            
    results = {k : v for k,v in results.items() if v > 0}
    results =  dict(sorted(results.items(), key = itemgetter(1), reverse = True))
    
    if as_dataframe:
        results =  pd.DataFrame(pd.Series(results)).reset_index()
        results.columns = ["Language", "Count"]
    
    return results
        
    
```

```python
#lets write a functiont o get all the scores for a set of languages

def sort_dict_by_values(d):
    return dict(sorted(d.items(), key = itemgetter(1), reverse = True))

def get_average_job_ratings_for_langs(df, langs_list):
    results = {}
    for l in langs_list:
        m = df[df['HaveWorkedLanguage'].str.contains(l+";") | df['HaveWorkedLanguage'].str.endswith(l)]["JobSatisfaction"].mean()
        results.update({l: m})
    
    return sort_dict_by_values(results)
    
```
